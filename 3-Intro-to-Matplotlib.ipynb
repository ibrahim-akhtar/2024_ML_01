## Matplotlib Example Workflow

```python
# 0. import matplotlib & get it ready

# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(6, 6)) # height & width pair

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Sample Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & Show
fig.savefig("sample-plot.png") # "images/sample-plot.png"
```

#### Making multiple plots at the same time (called subplots)

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))

# datas
x = np.random.randn(1000)

# PLot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(np.random.random(10), np.random.random(10));
ax4.hist(x);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# datas
x = np.random.randn(1000)

# PLot to each different axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(np.random.random(10), np.random.random(10));
ax[1, 1].hist(x);
```

### Which plotting to use? (pyplot vs. matplotlib OO method)

* When plotting something quickly, use pyplot (previous few ways)
* When plotting something more advanced, use OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method...
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method...
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
# pyplot method again with figsize
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10, 6));
```

```python
# OO method again
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim((45, 100));
```

```python
## OO method with scratch.... (above one was a little bit mixed)

fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease & Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.target.values
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,figsize=(10, 10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize ax0
ax0.set(title="Heart Disease & Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# customize ax1
ax1.set(title="Heart Disease & Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax1.axhline(over_50["thalach"].mean(),
          linestyle='--');


# add a title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease & Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"); #changes color scheme

# Customize the plot
ax.set(title="Heart Disease & Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# Customizing the limits of axes X and Y

# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,figsize=(10, 10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# customize ax0
ax0.set(title="Heart Disease & Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# changing x axis limits
ax0.set_xlim([50,80])
# changing y axis limits
ax0.set_ylim([100,600])

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(over_50["chol"].mean(),
          linestyle='--');


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")

# customize ax1
ax1.set(title="Heart Disease & Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# changing x axis limits
ax1.set_xlim([50,80])
# changing y axis limits
ax1.set_ylim([60,220])

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax1.axhline(over_50["thalach"].mean(),
          linestyle='--');


# add a title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
